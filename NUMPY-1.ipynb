# RANGE() =>
**when we want to use numbers in a particular range in Python**\
\
*SYNTAX ->*
`range(start, stop, step)`

```python
list(range(0, 5))
```

```python
list(range(0, 5, 2))
```

```python
list1 = [1, 3, 5, 7, 9]
list2 = [2, 4, 6, 8, 10]
```

```python
list1 * list2
```

```python
new_list = []

for i in range(0, len(list1)):
  x = list1[i] * list2[i]
  new_list.append(x)

print(new_list)
```
