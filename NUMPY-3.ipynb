# **ZEROS**


*   This function produces an array of zeros of whatever the size we want. In other words, the array we create will consist of all zeros.
*   The data type that we will write inside the zeros function is a tuple.
*   This tuple value is the size of the array that we want to create







```python
import numpy as np
```

```python
np.zeros(5)
```

```python
np.zeros((3, 5), dtype="int")
```
